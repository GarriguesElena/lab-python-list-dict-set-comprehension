````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]
inventory = {}

def initialize_inventory(products):
    inventory = {product: int(input(f"Enter the quantity of {product}s available: ")) for product in products}
    return inventory

inventory = initialize_inventory(products)
````

````python
print(inventory)
````

````python
customer_orders = {"hat", "mug", "t-shirt"}

def update_inventory(customer_orders, inventory):
    inventory = {product: inventory[product]- 1 if product in customer_orders else inventory[product] for product in inventory}
    inventory = {product: quantity for product, quantity in inventory.items() if quantity > 0}
            
    return inventory

inventory = update_inventory(customer_orders, inventory)
inventory
````

````python
def print_updated_inventory(inventory):
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")
        
print_updated_inventory(inventory)
````

````python
def get_customer_orders():
    x = int(input("How many items would you like to order?"))
    
    customer_orders = {input("Enter the product name: ") for x in range(x)}

    return customer_orders

get_customer_orders()
````

````python
def total_price(get_customer_orders):
    total_price = sum([int(input(f"Enter the price for {product}: ")) for product in customer_orders])

    return total_price
    
total_price(get_customer_orders)
````
